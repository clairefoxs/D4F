```python
df = pd.DataFrame.from_dict(data)
df_cities = pd.DataFrame.from_dict(data2)
```

```python
#Check for new entries without lat or long data. Update working df.
for x in df.itertuples():
    if x.lat == '': 
        new_lat = df_cities.lat[(df_cities.city == x.city) & (df_cities.state_name == x.state_name)].values[0]
        
        df.lat[(df.city == x.city) & (df.state_name == x.state_name)] = new_lat #assign new lat to working df
        
        sheet.update_cell(index+2, 6, new_lat) #update gsheet cell
    
    if x.lng == '':
        new_lng = df_cities.lng[(df_cities.city == x.city) & (df_cities.state_name == x.state_name)].values[0]
        
        df.lng[(df.city == x.city) & (df.state_name == x.state_name)] = new_lng #assign new lng to working df
        
        sheet.update_cell(index+2, 7, new_lng)\
            
```
